```python
#Identify the predictor variables and encode any string variables to equivalent integer codes:
df['gender'],_ = pd.factorize(df['gender'])
df.head(20)
df['NationalITy'],_ = pd.factorize(df['NationalITy'])
df.head(20)
df['PlaceofBirth'],_ = pd.factorize(df['PlaceofBirth'])
df.head(20)
df['StageID'],_ = pd.factorize(df['StageID'])
df.head(20)
df['GradeID'],_ = pd.factorize(df['GradeID'])
df.head(20)
df['SectionID'],_ = pd.factorize(df['SectionID'])
df.head(20)
df['Topic'],_ = pd.factorize(df['Topic'])
df.head(20)
df['Semester'],_ = pd.factorize(df['Semester'])
df.head(20)
df['Relation'],_ = pd.factorize(df['Relation'])
df.head(20)
df['ParentAnsweringSurvey'],_ = pd.factorize(df['ParentAnsweringSurvey'])
df.head(20)
df['ParentschoolSatisfaction'],_ = pd.factorize(df['ParentschoolSatisfaction'])
df.head(20)
df['StudentAbsenceDays'],_ = pd.factorize(df['StudentAbsenceDays'])
df.head(20)
df['Class'],_ = pd.factorize(df['Class'])
df.head(20)
```

```python
df.info()
```

```python
#Select the predictor feature and the target variable:
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
#train test split
from sklearn import model_selection
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# train the decision tree
from sklearn import tree
dtree = tree.DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=0)
dtree.fit(X_train, y_train)
```

```python
# use the model to make predictions with the test data
from sklearn import metrics
y_pred = dtree.predict(X_test)
# how did our model perform?
count_misclassified = (y_test != y_pred).sum()
print('Misclassified samples: {}'.format(count_misclassified))
count_classified = (y_test == y_pred).sum()
print('Correctly Classified Samples:{}'.format(count_classified))
accuracy = metrics.accuracy_score(y_test, y_pred)
print('Accuracy: {:.2f}'.format(accuracy))
```

```python
#Representation of an graph:
import graphviz
feature_names = X.columns

dot_data = tree.export_graphviz(dtree, out_file=None, filled=True, rounded=True,
                                feature_names=feature_names,  
                                class_names=class_names)
graph = graphviz.Source(dot_data)  
graph
```

```python
#Support Vector Machine:
from sklearn.svm import SVC
model = SVC()
model.fit(X_train,y_train)
```

```python
#Predictions:
pred = model.predict(X_test)
```

```python
#Model Evaluation:
from sklearn.metrics import classification_report,confusion_matrix
confusion_matrix(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```
